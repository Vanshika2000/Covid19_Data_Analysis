```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('covid19_Confirmed_dataset.csv')
print(df1.shape)
df1.head()
```

```python
df1.dropna(inplace = True, how = 'all' ,axis = 1)
df1.head()
```

```python
df1.loc[:, (df1 != 0).any(axis=0)]
df1.head()
```

```python
df1 = df1.drop(columns = ['Lat','Long'],axis = 1)
df1.head()
```

```python
df2 = df1.groupby(['Country/Region']).sum()
df2.head()
```

```python
df3 = df2.reset_index()
df3.head()
```

```python
df4 = df3.loc[df3['Country/Region']=='China']
df4
```

```python
df5 = df4.drop(columns = ['Country/Region'])
df5
```

```python
df6 = df5.transpose()
df7 = df6.reset_index()
df7.head()
df7.columns = ['DATE','COUNT']
df7.head()
```

```python
df7.plot(x = 'DATE', y= 'COUNT', kind ='line')
```

```python
sns.lineplot(x = 'DATE',y = 'COUNT', data = df7)
```

```python
A = []
for i in range(df2.shape[0]):
     A.append(df2.iloc[i,:].diff().max())
print(A)     
  
    
```

```python
df2['Max_Diff'] = A
```
